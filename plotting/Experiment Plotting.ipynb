```python
import numpy as np
import matplotlib.pyplot as plt

base_path = '../output/results/'
```

# Experiment Figure 5 LEFT

```python
exp_path = base_path + '/lr/Experiment Figure 5 (non-IID) - LEFT/'

fednova_left_train_loss = np.load(exp_path + 'model=lr_scheme=fednova_heteroE=False_heteroD=True_iid=False_dataset=synthetic_train_loss.npy')
fednova_left_test_loss = np.load(exp_path + 'model=lr_scheme=fednova_heteroE=False_heteroD=True_iid=False_dataset=synthetic_loss.npy')
fednova_left_test_accuracy = np.load(exp_path + 'model=lr_scheme=fednova_heteroE=False_heteroD=True_iid=False_dataset=synthetic_accuracy.npy')

fedavg_left_train_loss = np.load(exp_path + 'model=lr_scheme=fedavg_heteroE=False_heteroD=True_iid=False_dataset=synthetic_train_loss.npy')
fedavg_left_test_loss = np.load(exp_path + 'model=lr_scheme=fedavg_heteroE=False_heteroD=True_iid=False_dataset=synthetic_loss.npy')
fedavg_left_test_accuracy = np.load(exp_path + 'model=lr_scheme=fedavg_heteroE=False_heteroD=True_iid=False_dataset=synthetic_accuracy.npy')


fig, axs = plt.subplots(1, 3, figsize=(20, 5))

axs[0].plot(fednova_left_train_loss, label='FedNova', lw=5)
axs[0].plot(fedavg_left_train_loss, label='FedAvg', lw=5)
axs[0].set_title('Logistic Regression\nTraining Loss\n E = 5; C = 1.0', fontsize=24, fontweight='bold', fontstyle='italic')
axs[0].tick_params(axis='both', labelsize=18)
axs[0].legend(prop={'size': 18})
axs[0].set_xlabel('Communication Round', fontsize=20, fontweight='bold')
axs[0].set_ylabel('Loss Value', fontsize=20, fontweight='bold')
axs[0].grid()

axs[1].plot(fednova_left_test_loss, label='FedNova', lw=5)
axs[1].plot(fedavg_left_test_loss, label='FedAvg', lw=5)
axs[1].set_title('Logistic Regression\nTest Loss\n E = 5; C = 1.0', fontsize=24, fontweight='bold', fontstyle='italic')
axs[1].tick_params(axis='both', labelsize=18)
axs[1].legend(prop={'size': 18})
axs[1].set_xlabel('Communication Round', fontsize=20, fontweight='bold')
axs[1].set_ylabel('Loss Value', fontsize=20, fontweight='bold')
axs[1].grid()

axs[2].plot(fednova_left_test_accuracy, label='FedNova', lw=5)
axs[2].plot(fedavg_left_test_accuracy, label='FedAvg', lw=5)
axs[2].set_title('Logistic Regression\nTest Accuracy\n E = 5; C = 1.0', fontsize=24, fontweight='bold', fontstyle='italic')
axs[2].tick_params(axis='both', labelsize=18)
axs[2].legend(prop={'size': 18})
axs[2].set_xlabel('Communication Round', fontsize=20, fontweight='bold')
axs[2].set_ylabel('Accuracy [%]', fontsize=20, fontweight='bold')
axs[2].grid()

plt.subplots_adjust(wspace=0.3)
plt.show()
```
